```r
# 1. HW Function
# Calculates the expected Hardy-Weinberg genotype frequencies (AA, Aa, aa)
# from a vector of two allele frequencies (p and q).
HW <- function(allele_freq) {
  # Assuming biallelic: allele_freq[1] = p, allele_freq[2] = q
  p <- allele_freq[1]
  q <- allele_freq[2]

  # Calculate genotype frequencies: p^2, 2pq, q^2
  AA_freq <- p^2
  Aa_freq <- 2 * p * q
  aa_freq <- q^2

  genotype_freq <- c(AA_freq, Aa_freq, aa_freq)
  names(genotype_freq) <- c("AA", "Aa", "aa")

  return(genotype_freq)
}
```

```r
# 2. Main Simulation Function (Wright-Fisher Model with Drift)
# Runs multiple replicates for a given N, n_gen, and p_init, tracking p and H.
HW_Drift_Simulator <- function(N, n_gen, p_init, n_replicates) {
  # --- Initialization ---
  # Storage for all replicates
  all_p <- array(NA, dim = c(n_gen, 2, n_replicates))
  all_H_simulated <- matrix(NA, nrow = n_gen, ncol = n_replicates)

  p0 <- p_init
  q0 <- 1 - p0
  allele_freq_init <- c(p0, q0)
  H0 <- 2 * p0 * q0 # Initial Heterozygosity

  # Expected H Decay (Theoretical H_t = H0 * (1 - 1/(2N))^(t-1))
  H_expected <- H0 * (1 - 1 / (2 * N)) ^ (0:(n_gen - 1))

  set.seed(42) # Set a seed for reproducibility

  # --- Replicate Loop: Run the experiment multiple times ---
  for (r in 1:n_replicates) {
    pt_mat <- matrix(NA, nrow = n_gen, ncol = 2)
    pt_mat[1, ] <- allele_freq_init
    all_H_simulated[1, r] <- H0

    # --- Generation Loop ---
    for (t in 2:n_gen) {
      p_prev <- pt_mat[t - 1, 1]

      # Stop if fixation/loss occurred previously
      if (p_prev == 0 || p_prev == 1) {
        pt_mat[t, ] <- pt_mat[t - 1, ] # Maintain fixed frequency
        all_H_simulated[t, r] <- 0
        next
      }

      # Genetic Drift (Random sampling of 2N alleles)
      # This violates the infinite population size assumption (Question 1)
      total_alleles <- 2 * N
      n_A <- rbinom(1, size = total_alleles, prob = p_prev)
      p_next <- n_A / total_alleles

      # Store results for this generation
      pt_mat[t, ] <- c(p_next, 1 - p_next)
      all_H_simulated[t, r] <- 2 * p_next * (1 - p_next)
    }
    all_p[, , r] <- pt_mat
  }

  # --- Summary Calculation ---
  fixation_count <- sum(all_p[n_gen, 1, ] == 1)
  loss_count <- sum(all_p[n_gen, 1, ] == 0)

  return(list(
    N = N,
    H0 = H0,
    H_expected = H_expected,
    H_simulated_avg = rowMeans(all_H_simulated),
    all_p_mat = all_p,
    fixation_rate = fixation_count / n_replicates,
    loss_rate = loss_count / n_replicates
  ))
}
```

```r
# Define Simulation Parameters
n_gen <- 500         # Number of generations to run (Suggested: 500 or more)
p_init <- 0.5        # Initial allele frequency (Suggested: 0.01, 0.1, 0.5, 0.9)
n_replicates <- 50   # Number of independent runs for each N

# Define scenarios based on population size (N) - Suggested by assignment
N_scenarios <- c(100, 1000, 10000)

# Run Simulations for all scenarios
sim_results <- list()
for (N_val in N_scenarios) {
  cat(paste("Running simulation for N =", N_val, "...\n"))
  sim_results[[as.character(N_val)]] <- HW_Drift_Simulator(N = N_val,
                                                          n_gen = n_gen,
                                                          p_init = p_init,
                                                          n_replicates = n_replicates)
}
```

```r
# Output Summary Statistics (Used for answering Questions 1 and 4)
cat("\n\n--- Summary Statistics ---\n")
cat(paste("Parameters: Generations =", n_gen, ", Replicates =", n_replicates, ", Initial p =", p_init, "\n"))

for (N_val in N_scenarios) {
  results <- sim_results[[as.character(N_val)]]

  cat(paste("\nPopulation Size N =", N_val, "\n"))
  cat(paste("  Final Simulated H (Avg):", round(results$H_simulated_avg[n_gen], 4), "\n"))
  cat(paste("  Final Expected H (Theory):", round(results$H_expected[n_gen], 4), "\n"))
  cat(paste("  Prob. of Fixation (p=1):", round(results$fixation_rate, 4), "\n"))
  cat(paste("  Prob. of Loss (p=0):", round(results$loss_rate, 4), "\n"))
}
```

```r
par(mfrow = c(2, 1), mar = c(4, 4, 2, 1))

# Scenario 1: Small Population (N=100) - Strong Drift
plot(1:n_gen, sim_results[["100"]]$all_p_mat[, 1, 1], type = "l", ylim = c(0, 1),
     col = "blue", xlab = "Generation", ylab = "Allele Frequency (p)",
     main = paste("N=100 (Strong Genetic Drift) -", n_replicates, "Replicates"))
# Plot a few more replicates to show the random wandering
for (r in 2:5) {
  lines(1:n_gen, sim_results[["100"]]$all_p_mat[, 1, r], col = "lightblue")
}
abline(h = p_init, lty = 2, col = "gray")
```

```r
# Scenario 2: Large Population (N=10000) - Weak Drift
plot(1:n_gen, sim_results[["10000"]]$all_p_mat[, 1, 1], type = "l", ylim = c(0, 1),
     col = "darkgreen", xlab = "Generation", ylab = "Allele Frequency (p)",
     main = paste("N=10000 (Weak Genetic Drift) -", n_replicates, "Replicates"))
for (r in 2:5) {
  lines(1:n_gen, sim_results[["10000"]]$all_p_mat[, 1, r], col = "lightgreen")
}
abline(h = p_init, lty = 2, col = "gray")
```

```r
# Compares simulated vs. expected H for different N values (Questions 2 and 3)
par(mfrow = c(1, 1), mar = c(4.5, 4.5, 2, 1))

plot(1, type = "n", xlim = c(1, n_gen), ylim = c(0, sim_results[["100"]]$H0 * 1.05),
     xlab = "Generation", ylab = "Heterozygosity (H)",
     main = "Heterozygosity Decay due to Genetic Drift (Effect of N)")

colors <- c("red", "blue", "darkgreen")

# Add Expected Decay Curves (dashed lines)
for (i in 1:length(N_scenarios)) {
  N_val <- as.character(N_scenarios[i])
  # Expected decay is in dashed lines
  lines(1:n_gen, sim_results[[N_val]]$H_expected, col = colors[i], lty = 2, lwd = 1.5)
}

# Add Simulated Average Decay Curves (solid lines)
for (i in 1:length(N_scenarios)) {
  N_val <- as.character(N_scenarios[i])
  # Simulated average is in solid lines
  lines(1:n_gen, sim_results[[N_val]]$H_simulated_avg, col = colors[i], lty = 1, lwd = 2)
}

legend("topright",
       legend = c(
         paste("N=100 (Simulated)"),
         paste("N=1000 (Simulated)"),
         paste("N=10000 (Simulated)"),
         "Expected Decay (Theory)"
       ),
       col = c("red", "blue", "darkgreen", "black"),
       lty = c(1, 1, 1, 2), lwd = 2, bty = "n")
```
